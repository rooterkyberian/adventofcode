```python
import re
def read_stacks(stack_data):
    no_stacks = int(stack_data.split()[-1])
    stacks = [[] for _ in range(no_stacks)]
    for line in reversed(stack_data.split('\n')):
        for m in re.finditer(r"\[(.)\]", line):
            idx = m.start()//4
            stacks[idx].append(m.group(1))
    return stacks
stacks = read_stacks(stack_data)
from pprint import pprint
pprint(stacks)

print()

def do_moves(stacks: list[list[str]], moves: str):
    for line in moves.split('\n'):
        m = re.match(r"move (\d+) from (\d+) to (\d+)", line)
        count, src, dst  = map(int, m.groups())
        src = src - 1
        dst = dst - 1
        for _ in range(count):
            stacks[dst].append(stacks[src].pop())
    return stacks


def do_moves9001(stacks: list[list[str]], moves: str):
    for line in moves.split('\n'):
        m = re.match(r"move (\d+) from (\d+) to (\d+)", line)
        count, src, dst  = map(int, m.groups())
        src = src - 1
        dst = dst - 1
        stacks[dst].extend(stacks[src][-count:])
        stacks[src][-count:] = []
    return stacks

stacks = do_moves9001(stacks, moves)
pprint(stacks)

print(''.join(
' ' if not s else s[-1] for s in stacks
))
```
