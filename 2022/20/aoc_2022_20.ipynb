```python
from dataclasses import dataclass

data = open("input.txt").read()
```

```python
@dataclass
class N:
    v: int

    def __eq__(self, other):
        return super().__eq__(other) or self.v == other


def read_data(data):
    nums = [N(int(l)) for l in data.strip().split()]
    return nums


def example1_iter():
    d = """1 moves between 2 and -3:
2, 1, -3, 3, -2, 0, 4

2 moves between -3 and 3:
1, -3, 2, 3, -2, 0, 4

-3 moves between -2 and 0:
1, 2, 3, -2, -3, 0, 4

3 moves between 0 and 4:
1, 2, -2, -3, 0, 3, 4

-2 moves between 4 and 1:
1, 2, -3, 0, 3, 4, -2

0 does not move:
1, 2, -3, 0, 3, 4, -2

4 moves between -3 and 0:
1, 2, -3, 4, 0, 3, -2
    """.strip()
    for l in d.split("\n\n"):
        l = l.strip().split("\n")[-1]
        yield [int(e) for e in l.strip().split(", ")]


def _mixing(data):
    order = read_data(data)
    numbers = order[:]
    yield from __mixing(numbers, order)


def __mixing(numbers, order):
    for i in order:
        idx = numbers.index(i)
        numbers.pop(idx)
        idx += i.v
        idx %= len(numbers)
        if idx < 0:
            idx += len(numbers)
        if idx == 0:
            idx = len(order)
        numbers.insert(idx, i)
        yield numbers


def mixing(numbers, order):
    for x in __mixing(numbers, order):
        pass
    return x


def task1(data, mix=1, mul=1):
    order = read_data(data)
    order = [N(n.v * mul) for n in order]
    numbers = order[:]
    for i in range(mix):
        print(f"mixed {i}")
        numbers = mixing(numbers, order)
    zero_idx = next(c for c, n in enumerate(numbers) if n.v == 0)
    values = []
    for offset in (1000, 2000, 3000):
        n = numbers[(zero_idx + offset) % len(numbers)]
        values.append(n.v)
    print(values)
    return sum(values)


def task2(data):
    return task1(data, 10, 811589153)


# for i, (n, a, b) in enumerate(zip([1, 2, -3, 3, -2, 0, 4], example1_iter(), _mixing(example1))):
#    print(n, a, b, a==b)

print(f"{task1(example1)=}")  # 3
print(f"{task1(data)=}")  # 988
print(f"{task2(example1)=}")
print(f"{task2(data)=}")
```
