```python
def cpu1(data):
    X = 1
    c = 0

    def cycle():
        nonlocal c
        c = c + 1
        yield c, X

    for line in data.strip().split("\n"):
        if line == "noop":
            yield from cycle()
        if line.startswith("addx"):
            cmd, v = line.split()
            yield from cycle()
            yield from cycle()
            X = X + int(v)
    yield from cycle()


def task1(data):
    s = 0
    for c, X in cpu1(data):
        if c in range(20, 221, 40):
            s = s + c * X
    return s


task1(data)
```

```python
from io import StringIO


def task2(data):
    screen = StringIO()
    for c, X in cpu1(data):
        i = (c - 1) % 40
        if i in range(X - 1, X + 2):
            screen.write("#")
        else:
            screen.write(".")
        if i == 39:
            screen.write("\n")
    print(screen.getvalue())


task2(data)
```
