```python
import itertools

data = open("input.txt").read().strip()
```

```python
example = """
R 4
U 4
L 3
D 1
R 4
D 1
L 5
R 2
""".strip()
```

```python
def task1(data):
    visited = set()
    hx = hy = tx = ty = 0
    for move in data.split("\n"):
        d, step = move.split()
        step = int(step)
        for _ in range(step):
            c = 1
            if d == "U":
                hy = hy - c
            elif d == "D":
                hy = hy + c
            elif d == "L":
                hx = hx - c
            elif d == "R":
                hx = hx + c

            x_delta = hx - tx
            y_delta = hy - ty
            if abs(x_delta) > 1:
                ty = hy
                tx = hx - 1 if tx < hx else hx + 1
            elif abs(hy - ty) > 1:
                tx = hx
                ty = hy - 1 if ty < hy else hy + 1
            # print((tx,ty))

            visited.add((tx, ty))

    return len(visited)


task1(data)
```

```python
def simulate_step(hx, hy, tx, ty):
    x_delta = hx - tx
    y_delta = hy - ty
    if abs(x_delta) > 1 and abs(x_delta) == abs(y_delta):
        ty = hy - 1 if y_delta > 0 else hy + 1
        tx = hx - 1 if x_delta > 0 else hx + 1
    if abs(x_delta) > 1 and abs(x_delta) > abs(y_delta):
        ty = hy
        tx = hx - 1 if x_delta > 0 else hx + 1
    if abs(y_delta) > 1 and abs(y_delta) > abs(x_delta):
        tx = hx
        ty = hy - 1 if y_delta > 0 else hy + 1

    return tx, ty


def pairwise(iterable):
    a, b = itertools.tee(iterable)
    next(b, None)
    return zip(a, b)


def pknotes(knotes):
    return " ".join(f"{i}:{k[0]},{k[1]}" for i, k in enumerate(knotes))


def task2(data, n=10, *, v=0):
    visited = set()
    knotes = [[0, 0] for _ in range(n)]
    for move in data.split("\n"):
        if v > 0:
            print(move)
        d, step = move.split()
        step = int(step)
        for s in range(step):
            hx, hy = knotes[0]
            c = 1
            if d == "U":
                hy = hy - c
            elif d == "D":
                hy = hy + c
            elif d == "L":
                hx = hx - c
            elif d == "R":
                hx = hx + c

            knotes[0] = [hx, hy]

            for i in range(1, len(knotes)):
                r = list(simulate_step(*knotes[i - 1], *knotes[i]))
                if v > 2:
                    print(f"   k{i} {knotes[i-1]} {knotes[i]} = {r}")
                knotes[i] = r
                # print("   k", i, knotes[i-1], knotes[i])
            visited.add(tuple(knotes[-1]))
            if v > 1:
                print(f"  s{s}", pknotes(knotes))
        if v > 0:
            print(pknotes(knotes))

    return len(visited)


task2(example, 10, v=3)
```
